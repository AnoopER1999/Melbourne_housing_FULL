```python
df.dropna(inplace=True)
```

```python
df
```

```python
import statsmodels.formula.api as smf
model = smf.ols(formula='Price~Rooms+Distance+Postcode+Bathroom+Landsize+YearBuilt+Latitude+Longtitude+Car',data=df).fit()
model.summary()
```

```python
y= df[['Price']]
X = df[['Rooms','Distance','Postcode','Bathroom','Landsize','YearBuilt','Latitude','Longtitude','Propertycount','Car']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
y_pre = pd.DataFrame(model.predict(X))
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_pre)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,y_pre)
```

```python
model.score(X,y)
```
